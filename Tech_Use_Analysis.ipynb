```python
import numpy as np
import statsmodels.api as sm

X = df[['age', 'sleep_duration_hours', 'stress_level', 'physical_activity_hours_per_week', 'caffeine_intake_mg_per_day']]
Y = df[['daily_screen_time_hours']]

X = sm.add_constant(X)

model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```
